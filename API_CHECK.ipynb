```python
import os
from dotenv import load_dotenv
from langchain_nvidia_ai_endpoints import ChatNVIDIA

load_dotenv()  # reads .env and loads variables into os.environ

client = ChatNVIDIA(
    model="moonshotai/kimi-k2.6",
    api_key=os.getenv("KIMI_API_KEYK27"),
    temperature=1,
    top_p=1,
    max_completion_tokens=16384,
)

lc_messages = [{"role": "user", "content": "Hello!"}]
response = client.invoke(lc_messages)

if response.additional_kwargs and "reasoning_content" in response.additional_kwargs:
    print(response.additional_kwargs["reasoning_content"])
print(response.content)
```
